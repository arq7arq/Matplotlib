MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x = [0,1,2,3,4,5]
y = [i * 2  for i in x]

# resize the graph
plt.figure(figsize=(4,3), dpi=100)


# create the plot
plt.plot(x,y, label='2x', color='cyan', marker='.', markersize=10, markeredgecolor='blue', linestyle='--')


# number two graph
x2 = np.arange(0,4,0.5)
plt.plot(x2,x2 ** 2, color='red', label='X^2')


# adding title
plt.title('my first graph', fontdict={'fontname': 'Comic Sans MS','fontsize':20})

# adding x ,y labels
plt.xlabel('x')
plt.ylabel('y')

# adding the plot label to plot
plt.legend()

# saveing graph
plt.savefig('test.png',dpi=300)

# showing the graph
plt.show()
```

## Bar Chart

```python
labels = ['M','S','A']
values = [1,28,7]

bars = plt.bar(labels, values)

patterns = ['/','O','+']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

## Line Graph

```python
gas = pd.read_csv('./gas_price.csv')
gas.head()
```

```python
plt.figure(figsize=(8,5))

plt.title('gas prices in USD')

for country in gas.columns[1:]:
    plt.plot(gas['Year'], gas[country], label = country, marker='.')
    # print(country)

plt.xticks(gas['Year'][::3])

plt.xlabel('Years')
plt.ylabel('USD')

plt.legend()
plt.show()
```

## Histograms

```python
fifa = pd.read_csv('./fifa_data.csv')
fifa.head()
```

```python
bins = [x*10 for x in range(4,11)]
plt.hist(fifa['Overall'],bins=bins, color='#abcdef')

plt.xticks(bins)
plt.ylabel('Number of players')
plt.xlabel('Overall of player')
plt.title('Distribution of overall players')

plt.show()
```

## Pie Chart

```python
left = fifa[fifa['Preferred Foot'] == 'Left']['ID'].count()
right = fifa[fifa['Preferred Foot'] == 'Right']['ID'].count()

labels = ['left', 'right']
colors = ['#abcdef','#012345']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.1f %%')
plt.title('Foot Preference of players')

plt.show()
```

## Boxplot

```python
plt.style.use('default')

barcelona = fifa[fifa['Club'] == 'FC Barcelona']['Overall']
real = fifa[fifa['Club'] == 'Real Madrid']['Overall']

plt.boxplot([real,barcelona])

plt.show()
```
